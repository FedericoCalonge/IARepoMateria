### Ejemplo básico de manejo de Datos

```python
#*********************************************************************************
# El siguiente ejemplo se encuentra basado en la documentación oficial 
# y en los ejemplos de:
# 1) https://towardsdatascience.com/linear-regression-using-python-ce21aa90ade6
# 2) https://machinelearningmastery.com/machine-learning-in-python-step-by-step/
#**********************************************************************************
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   #Data visualisation libraries 
import seaborn as sns
```

### Carga del Dataset

```python
# Seaborn is a Python data visualization library based on matplotlib. 
# It provides a high-level interface for drawing attractive and informative statistical graphics.

# Load dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset_iris = pd.read_csv(url, names=names)# Resulta buena práctica, analizar los datos

# Resulta importante conocer las dimensiones del Dataset
print(dataset_iris.shape)
```

```python
dataset_iris.head(10) #Observamos los 10 primeros registros
```

### Análisis estadístico de los datos

```python
# Análisis Estadístico
print(dataset_iris.describe())
```

```python
#Grafico de Dispersion para una sola variable (sepal-width) y asi analizamos lo de arriba:
sns.distplot(dataset_iris['sepal-width'])  #Vemos que sigue una distribución NORMAL.
```

```python
#Gráficos de Histograma y Dispersión
# Los Histogramas muestran la información con respecto a una misma variable.
# Los gráficos de Dispersión muestran la relación entre dos variables.
sns.pairplot(dataset_iris)
```
